```python
#import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
df = pd.read_csv('House Prices1.csv')
X = df[['Area(sq ft)', 'Bedrooms', 'Age(years)']]
y = df['Prices($)']
```

```python
#tesplit the data into 80 and 20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#import model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2*100}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Fit')


plt.title('Predicted Price vs Actual Price')
plt.xlabel('Actual Price ($)')
plt.ylabel('Predicted Price ($)')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Step 1: Normalize the dataset using StandardScaler
scaler = StandardScaler()
```

```python
# Fit and transform the training data
X_train_normalized = scaler.fit_transform(X_train)
```

```python

# Transform the test data (using the same scaler to avoid data leakage)
X_test_normalized = scaler.transform(X_test)
```

```python

# Step 2: Train the model on the normalized data
model_normalized = LinearRegression()
model_normalized.fit(X_train_normalized, y_train)
```

```python

# Step 3: Make predictions with the normalized model
y_pred_normalized = model_normalized.predict(X_test_normalized)
```

```python
# Step 4: Compare results with the unnormalized model
# Predictions with the unnormalized model
y_pred_unnormalized = model.predict(X_test)
```

```python

# Step 5: Evaluate both models
mse_normalized = mean_squared_error(y_test, y_pred_normalized)
mse_unnormalized = mean_squared_error(y_test, y_pred_unnormalized)
```

```python
r2_normalized = r2_score(y_test, y_pred_normalized)
r2_unnormalized = r2_score(y_test, y_pred_unnormalized)

```

```python
print("Results for Normalized Data:")
print(f"Mean Squared Error: {mse_normalized}")
print(f"R-squared: {r2_normalized}")

print("\nResults for Unnormalized Data:")
print(f"Mean Squared Error: {mse_unnormalized}")
print(f"R-squared: {r2_unnormalized}")
```
